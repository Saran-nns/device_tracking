##### TREND 

###### Trend is set by differencing parameter: d 
A model with no orders of differencing assumes that the original series is stationary (among other things, mean-reverting). 
A model with 1 order of differencing assumes that the original series has a constant average trend 
(e.g. a random walk or SES-type model, with or without growth). A model with 2 orders of 
total differencing assumes that the original series has a time-varying trend (e.g. a random trend or LES-type model).
